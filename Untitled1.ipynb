```python
# importing the module
import timeit
import numpy as np
import sys
# sample function that returns square
# of the value passed
def print_square(x):
    return (x**2)
  
    
def find_linear(A,x): # A = array , x = key to search
    pos = -1
    i = 0
    while (A[i] !=x and i < len(A)-1):
        i = i + 1
    if (A[i] == x):
        pos = i
    return pos


def find_binary(A,x): # A = array , x = key to search
    pos = -1
    L = 0
    U = len(A) - 1 
    M = int((L + U) / 2)
    while(A[M] != x and L<U ):
        if x > A[M] :
            L = M + 1
        else:
            U = M - 1
        
        M = int((L + U) / 2)
    if (A[M] == x):
        pos = M
    return pos


def find_linear_rec(A,x,l,r): # A = array , x = key to search
    sys.setrecursionlimit(10000000)
    if l < r: 
        return -1
    if A[l] == x: 
        return l 
    if A[r] == x: 
        return r 
    return find_linear_rec(A,x,l+1,r-1)


def find_binary_rec(A,x,L,U):
    M = int((L+U)/2)
    if (A[M] == x) :
        return M
    else:
        if (L>U):
            return -1
        else:
            if (x > A[M] ):
                return find_binary_rec(A,x,M+1,U)
            else:
                return find_binary_rec(A,x,L,M-1)

            
def find_avg(list):
    sum = 0
    for e in list:
        sum += e 
    avg = sum / int(len(list))
    return avg


N = 10000000


Arr = np.random.randint(low=0, high=N, size=(N,))


print("Unsort : " + str(Arr))


Arr.sort()
print("Sorted : " + str(Arr))



key = np.random.randint(low=0, high=N)


t_linear = timeit.repeat(lambda: find_linear(Arr,key), number=10, repeat=5)


t_binary = timeit.repeat(lambda: find_binary(Arr,key), number=10, repeat=5)


t_linear_rec = timeit.repeat(lambda: find_linear_rec(Arr,key,0,len(Arr)-1), number=10, repeat=5)


t_binary_rec = timeit.repeat(lambda: find_binary_rec(Arr,key,0,len(Arr)-1), number=10, repeat=5)



print("Elapsed time of linear : " + str(t_linear))
print("Elapsed time of binary : " + str(t_binary))
print("Elapsed time of recursive linear : " + str(t_linear_rec))
print("Elapsed time of recursive binary : " + str(t_binary_rec))

print("คำตอบที่ได้จากการหาค่าเฉลี่ย")
print(" average time of linear :" + str(find_avg(t_linear)))
print(" average time of binary :" ,find_avg(t_binary))
print(" average time of linear_recursive :" ,find_avg(t_linear_rec))
print(" average time of bianry_recursive :" ,find_avg(t_binary_rec))
```
